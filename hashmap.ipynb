```python
#Make a class called Node that will contain the data in the linked list
class Node:
   # Privately initialize the function with instance/s and key and val parameters
   def __init__(self, key, val):
      # Create an instance of key
      self.key = key
      # Create an instance of val
      self.val = val
      # There is initially no next node
      self.next = None
# Create a class called LinkedList
class LinkedList:
   # Initialize the prehead with instance/s
   def __init__(self):
      # Starting point of the linked list which starts empty
      self.prehead = Node(None, None)
   # Create a function to search through the linked list using instance/s and key parameter
   def search(self, key):
      # p is variable that goes to the next node
      p = self.prehead.next
      # while loop of p that keeps going until key is found or is not
      while p:
         # returns the value if it is in the linked list
         if p.key == key:
            return p
         # if not, it goes to the next node
         p = p.next
      # none is returned if p is not a key to the list
      return None
   # Create a add function that adds a node with instance/s, key parameter, and value parameter
   def add(self, key, val):
      # p is a variable that uses the search function previously defined
      p = self.search(key)
      # if p, it becomes a value parameter
      if p:
         p.val = val
      # if anything else besides p
      else:
         # the node is created from the node class with key and value parameters
         node = Node(key, val)
         # the node goes to the next prehead
         self.prehead.next, node.next = node, self.prehead.next
   # Create a get function with instance/s and key parameter
   def get(self, key):
      # p is a variable that uses the search function previously defined
      p = self.search(key)
      # if p, it returns the value of p
      if p:
         return p.val
      # if anything else besides p
      else:
         #Nothing is returned
         return None
   # Create a remove function with instance/s and a key
   def remove(self, key):
      # Create a prev variable with the prehead to go from the previous node
      prev = self.prehead
      # The current value which is cur variable is the value after the prev value
      cur = prev.next
      # if the current variable is the key that you entered to remove, it breaks
      while cur:
         if cur.key == key:
            break
         prev, cur = cur, cur.next
      # if the current variable, the next variable equals the previous variable?
      if cur:
         prev.next = cur.next
   # Creates a serialize function with instance/s
   def serialize(self):
      # p is a variable that with the prehead goes to the next node
      p = self.prehead.next
      # ret is a blank list
      ret = []
      # if p, it returns the value of p
      while p:
         # add the key and value to the list
         ret.append([p.key, p.val])
         # go to the next set
         p = p.next
      # return the list created
      return ret
# Creates a class called myhashmap
class MyHashMap:
   # Initialize the prehead with instance/s
   def __init__(self):
      self.size = 1033
      self.arr = [LinkedList() for _ in range(self.size)]
   def _hash(self, key):
      return key % self.size
   def put(self, key, value):
      h = self._hash(key)
      self.arr[h].add(key, value)
   def get(self, key):
      h = self._hash(key)
      ret = self.arr[h].get(key)
      if ret is not None:
         return ret
      else:
         return -1
   def remove(self, key):
      h = self._hash(key)
      self.arr[h].remove(key)
```
