```python
# Reverse Operation
"""
Algorithm
Step by step process to reverse a linked list is as follows âˆ’

1 START
2. We use three pointers to perform the reversing: prev, next, head.
3. Point the current node to head and assign its next value to the prev node.
4. Iteratively repeat the step 3 for all the nodes in the list.
5. Assign head to the prev node.
"""
#Make a class called Node that will contain the data in the linked list
class Node:
   # Privately initialize the function with instance/s and blank data parameter in the linked list
   def __init__(self, data=None):
      # Create an instance of data
      self.data = data
      # There is initially no next node
      self.next = None

# Make a class called SLL that will contain the head for the linked list and the listprint function
class SLL:
   # Privately initialize the function with a blank head to start the linked list with instance/s
   def __init__(self):
      # There is initially no head
      self.head = None

   # Create a function that prints the linked list with instance/s
   def listprint(self):
      # Create a variable to represent the head of the linked list
      printval = self.head
      print("Linked List: ")
      # If there is data to print, the head will not be None and creates a while loop to go through the linked list
      while printval is not None:
         # Prints the current node in the linked list
         print (printval.data)
         # Goes to the next node in the linked list
         printval = printval.next
   # Creat a function that reverses the linked list order with instance/s
   def reverse(self):
      # Prev variable is none to later add
      prev = None
      # Curr variable starts at the head
      curr = self.head
      # Creates a while loop
      while(curr is not None):
         # Define variable called next (Confusing)
         next = curr.next
         curr.next = prev
         prev = curr
         curr = next
      self.head = prev

# Creates a list using the SLL class
l1 = SLL()
# Head for the linked list
l1.head = Node("731")
# Data for the first node
e2 = Node("672")
# Data for the second node
e3 = Node("63")

# Go from the head to the first node
l1.head.next = e2
# Go from the first node to the second node
e2.next = e3

# Print the linked list
l1.listprint()
# Reverse the linked list
l1.reverse()
# Print the reversed linked list
print("After reversing: ")
l1.listprint()
```
